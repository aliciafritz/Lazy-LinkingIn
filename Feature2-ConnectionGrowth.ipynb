```python
import pandas as pd
import matplotlib.pyplot as plt

def connections_growth_graph (csv_file_path):
    
    # Read the CSV file into a pandas DataFrame
    df2 = pd.read_csv(csv_file_path)
    print (df2)

    # Convert the 'dates' column to datetime format
    df2['dates'] = pd.to_datetime(df2['dates'])

    # Group the data by date and count the number of connections for each date
    connections_over_time = df2.groupby('dates').size().reset_index(name='Connections')
    
    return connections_over_time
```

```python
# calling the connections_over_time function

connectionsgrowth = connections_growth_graph ('connections2.csv')


# Plot the graph
plt.figure(figsize=(10, 6))
plt.plot(connectionsgrowth['dates'], connectionsgrowth['dates'], marker='o')
plt.title('LinkedIn Connections Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Connections')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
